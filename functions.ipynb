```python
def average_yield(yields):
  total=sum(yields)
  return total/len(yields)
crop_yields=[1200,1350,1100,1500]
avg=average_yield(crop_yields)
print("average_yield:",avg)
```

```python
def increase_yield(yields,increament):
  for i in range(len(yields)):
    yields[i]+=increament
crop_yields=[1200,1350,1100]
increase_yield(crop_yields,100)
print(crop_yields)
```

```python
def convert_to_quintals(yields):
  result=[]
  for value in yields:
    result.append(value/100)
  return result
crop_yields=[1200,1350,1100]
quintals=convert_to_quintals(crop_yields)
print(quintals)
```

```python
def high_yield_crops(yields,threshold):
  high=[]
  for value in yields:
    if value==threshold:
      high.append(value)
      return high
crop_yields=[1200,1000,1500,800]
print(high_yield_crops(crop_yields,1000))
```

```python
def sort_prices(prices):
  prices.sort()
  return prices
crop_prices=[1800,1500,2200,2000]
print(sort_prices(crop_prices))
```

```python
def sort_prices(prices):
  prices.sort(reverse=True)
  return prices
crop_prices=[1800,1500,2200,2000]
print(sort_prices(crop_prices))
```

```python
def sort_prices(prices):
  prices.sort()
  crop_prices_sorted=sorted(prices)
  return crop_prices_sorted
crop_prices=[1800,1500,2200,2000]
list1=sort_prices(crop_prices)
print(list1)

```

```python
def collect_yields():
  yields=[]
  n=int(input("enter the number of fields:"))
  for i in range(n):
    value=int(input("enter yield:"))
    yields.append(value)
  return yields
data=collect_yields()
print("collected yields:",data)
```

```python
def yield_statistics(yields):
  print("maximum yield:",max(yields))
  print("minimum yield:",min(yields))
  print("total yields:",len(yields))
crop_yields=[1200,1350,1100]
yield_statistics(crop_yields)
```

```python
def display_yields(yields_list):
  print("\nrecorded crop yields(in kg):")
  for i in range(len(yields_list)):
     print(f"field{i+1}:{yields_list[i]}kg")
crop_yields=[1200,1350,1100]
display_yields(crop_yields)
```
