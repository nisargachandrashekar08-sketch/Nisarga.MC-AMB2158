```python
import matplotlib.pyplot as plt
data=[
    30,37,29,95,24,65,2,4,25,35,35,26,8,9,25,24,47,43,57,21,14
]
plt.figure(figsize=(8,5))
plt.hist(
    data,bins=5,color="blue",
    edgecolor='black',
    alpha=0.5,density=True,
    range=(30,80)
)
plt.xlabel("value")
plt.ylabel("density")
plt.title("Histogram without Random data")
plt.grid(True)
plt.tight_layout()
plt.show()
```
