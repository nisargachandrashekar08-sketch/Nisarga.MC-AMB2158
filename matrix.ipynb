```python
#Transpose
import numpy as np
matrix=[[1,2],
   [3,4]]
transposed=np.transpose(matrix)
print(transposed)
```

```python
#matrix determinant(2*2)
import numpy as np
A=[[1,2],
   [3,4]]
det=np.linalg.det(A)
print(det)
```

```python
#(3*3)
import numpy as np
A=[[1,2,3],
   [7,4,5],
   [9,6,7]]
det=np.linalg.det(A)
print(det)
```

```python
#matrix inverse
import numpy as np
A=[[1,2],
   [3,4]]
inverse=np.linalg.inv(A)
print(inverse)
```

```python
#matrix ranking
import numpy as np
A=[[1,2],
   [3,4]]
rank=np.linalg.matrix_rank(A)
print(rank)
```

```python
#eigenvalues and eigenvectors
import numpy as np
A=[[1,2],
   [3,4]]
eigenvalues,eigenvectors=np.linalg.eig(A)
print(eigenvalues,eigenvectors)
```

```python
#solving linear system
import numpy as np
A=[[1,2],
 [3,4]]
B=[[5,6],
   [7,8]]
Ax=B
solution=np.linalg.solve(A,B)
print(solution)
```
