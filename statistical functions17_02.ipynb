```python
import pandas as pd
import numpy as np
#create sample dataframe
df=pd.DataFrame({
    'name':['alice','bob','charlie','david','eva','frank'],
    'department':['HR','IT','IT','Finance','HR','Finance'],
    'salary':[50000,70000,70000,60000,52000,60000],
    'experience':[2,5,4,6,3,6]
})
print(df)
```

```python
#sort by one column
import pandas as pd
import numpy as np

df=pd.DataFrame({
    'name':['alice','bob','charlie','david','eva','frank'],
    'department':['HR','IT','IT','Finance','HR','Finance'],
    'salary':[50000,70000,70000,60000,52000,60000],
    'experience':[2,5,4,6,3,6]
})
sal=df.sort_values('salary')
print(sal)
```

```python
#decending order
import pandas as pd
import numpy as np

df=pd.DataFrame({
    'name':['alice','bob','charlie','david','eva','frank'],
    'department':['HR','IT','IT','Finance','HR','Finance'],
    'salary':[50000,70000,70000,60000,52000,60000],
    'experience':[2,5,4,6,3,6]
})
sal=df.sort_values(['salary','experience'],ascending=[False,True])
print(sal)
```

```python
#sort by index
import pandas as pd
import numpy as np

df=pd.DataFrame({
    'name':['alice','bob','charlie','david','eva','frank'],
    'department':['HR','IT','IT','Finance','HR','Finance'],
    'salary':[50000,70000,70000,60000,52000,60000],
    'experience':[2,5,4,6,3,6]
})
des=df.sort_index(ascending=False)
print(des)
```

```python
#basics ranking(default ascending)
import pandas as pd
import numpy as np

df=pd.DataFrame({
    'name':['alice','bob','charlie','david','eva','frank'],
    'department':['HR','IT','IT','Finance','HR','Finance'],
    'salary':[50000,70000,70000,60000,52000,60000],
    'experience':[2,5,4,6,3,6]
})
df['rank_salary']=df['salary'].rank()
print(df)
```

```python
#basics ranking(decending)
import pandas as pd
import numpy as np

df=pd.DataFrame({
    'name':['alice','bob','charlie','david','eva','frank'],
    'department':['HR','IT','IT','Finance','HR','Finance'],
    'salary':[50000,70000,70000,60000,52000,60000],
    'experience':[2,5,4,6,3,6]
})
df['rank_salary']=df['salary'].rank(ascending=False)
print(df)
```

```python
#dense ranking(highest salary ranking 1)
import pandas as pd
import numpy as np

df=pd.DataFrame({
    'name':['alice','bob','charlie','david','eva','frank'],
    'department':['HR','IT','IT','Finance','HR','Finance'],
    'salary':[50000,70000,70000,60000,52000,60000],
    'experience':[2,5,4,6,3,6]
})
df['dense_rank_salary']=df['salary'].rank(ascending=False,method='dense')
print(df)
```

```python
#value counts example(frequency)
df=df['department'].value_counts()
print(df)
```

```python
#proportion(percentage,distribution)
import pandas as pd
import numpy as np

df=pd.DataFrame({
    'name':['alice','bob','charlie','david','eva','frank'],
    'department':['HR','IT','IT','Finance','HR','Finance'],
    'salary':[50000,70000,70000,60000,52000,60000],
    'experience':[2,5,4,6,3,6]
})
df=df['department'].value_counts()
print(df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.DataFrame({
    'name':['alice','bob','charlie','david','eva','frank'],
    'department':['HR','IT','IT','Finance','HR','Finance'],
    'salary':[50000,70000,70000,60000,52000,60000],
    'experience':[2,5,4,6,3,6]
})
#line plot
df[['salary','experience']].plot()
plt.show()
```

```python
#bar chart
df.plot(kind='bar',x='name',y='salary')
plt.show()
```

```python
#histogram
df['salary'].plot(kind='hist',bins=5)
plt.show()
```

```python
#scatter plot
df.plot(kind='scatter',x='experience',y='salary')
plt.show()
```

```python
#box plot
df[['salary','experience']].plot(kind='box')
plt.show()
```
